```python
svm.intercept_
```

## Przyk≈Çad 3

```python
X,y= make_blobs(n_samples=100, centers=2,
                random_state=0, cluster_std=1.2)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=50)
```

```python
svm=SVC(kernel='rbf')
svm.fit(X,y)
```

```python
y_pred=svm.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred, s=50)
```

```python
svm=SVC(kernel='rbf', C=1)
svm.fit(X,y)
```

```python
y_pred=svm.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred, s=50)
```

```python
def svm_result(C):
  svm=SVC(kernel='rbf', C=C)
  svm.fit(X,y)
  y_pred=svm.predict(X)
  plt.scatter(X[:,0], X[:,1], c=y_pred, s=50)
  plt.title(f'C={C}')
```

```python
svm_result(10)
```

```python
svm_result(0.001)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
wine=pd.read_csv('ML-datasets/wine.csv')
```

```python
y=wine.pop('Class').astype('category').cat.codes
```

```python
X=wine
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred=knn.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
knn=KNeighborsClassifier(n_neighbors=10, metric='chebyshev')
knn.fit(X_train, y_train)
y_pred=knn.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=[]

for i in range(1,40):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  y_pred=knn.predict(X_test)
  acc=accuracy_score(y_test, y_pred)
  accuracy.append(acc)
```

```python
accuracy
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40), accuracy, color='red',
         linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Accuracy w zale≈ºno≈õci od liczby sƒÖsiad√≥w')
plt.xlabel('k- liczba sƒÖsiad√≥w')
plt.ylabel('accuracy')
```

```python
params={'metric':['euclidean', 'manhattan', 'chebyshev'], 'n_neighbors':[3,5,10,15], 'weights':['distance', 'uniform']}
```

```python
knn=KNeighborsClassifier()
searcher=GridSearchCV(knn, params)
searcher.fit(X, y)
```

```python
searcher.best_params_
```

```python
searcher.best_score_
```

```python
knn=KNeighborsClassifier(metric='manhattan', n_neighbors=15, weights='distance')
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
